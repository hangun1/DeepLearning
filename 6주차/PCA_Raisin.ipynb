```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Raisin_Dataset.csv')
df.head()
```

```python
features = df.drop(columns=['Class'])
labels = df['Class']
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
# PCA 차원 수 2
pca_2 = PCA(n_components=2)
pca_features_2 = pca_2.fit_transform(scaled_features)
explained_variance_2 = pca_2.explained_variance_ratio_
```

```python
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(pca_features_2, labels, test_size=0.2, random_state=42)
clf_2 = RandomForestClassifier(random_state=42)
clf_2.fit(X_train_2, y_train_2)
y_pred_2 = clf_2.predict(X_test_2)
accuracy_2 = accuracy_score(y_test_2, y_pred_2)
```

```python
print("\nPCA with n_components=2")
print(f"Explained Variance Ratio: {explained_variance_2}")
print(f"Accuracy: {accuracy_2:.2f}")
print("Classification Report:")
print(classification_report(y_test_2, y_pred_2))
```

```python
plt.figure(figsize=(7, 4))
plt.plot(range(1, len(explained_variance_2) + 1), explained_variance_2, marker='o', label='n_components=2')
plt.title('Explained Variance Ratio by PCA Components (n_components=2)')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.legend()
plt.show()
```
